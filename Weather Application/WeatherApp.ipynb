```python
#Course: DSC510
#Assignment 12.1
#Date:05/31/2020
#Description: Programming Final

import requests
import json

api_key="d633a70ca405fbbe8733a0a5bafef70c" #saves api key from OpenWeather website

base_url="http://api.openweathermap.org/data/2.5/weather?" #saves base url

def main_function():
    prompt=input("Please enter your city name or zip code.\n") #prompt for users
    while True:
        if prompt.isdigit(): #separates zip code in responses
            zip_function(prompt)

        else: #separates strings in responses
            city_function(prompt)

def zip_function(prompt):
    complete_url_zip=base_url+"zip="+prompt+"&appid="+api_key+"&units=imperial" #adds in url information to get to zip code api access
    try:
        response_zip=requests.get(complete_url_zip) #gets information from api
    except:
        print('Connection was not successful.') #notifies user if cannot connect to api
        exit()
    if response_zip.status_code==404:
        print("Sorry. That is not a valid input. Please try again.") #notifies user if their input was not valid
        main_function()
    print("Today's weather forecast for ",prompt,":")
    json_response_zip=response_zip.json() #converts from json
    temp=json_response_zip["main"]["temp"] #gets current temperature
    print("Current temperature:",temp,"F")
    high=json_response_zip["main"]["temp_max"] #gets high temperature
    print("High Temperature:",high,"F")
    low=json_response_zip["main"]["temp_min"] #gets low temperature
    print("Low Temperature:",low,"F")
    humidity=json_response_zip['main']['humidity'] #gets humidity
    print("Humidity: ",humidity,"%")
    weather_condition=json_response_zip["weather"][0]["description"] #gets weather condition
    print("Weather Condition: ",weather_condition)
    repeat_function()

def city_function(prompt):
    complete_url_city=base_url+"q="+prompt+"&appid="+api_key+"&units=imperial" #puts in info for url to get api by city
    try:
        response_city=requests.get(complete_url_city) #gets full url
    except:
        print("Connection was not successful.") #notifies user if unable to connect to api
        exit()
    if response_city.status_code==404: #tests if invalid response
        print("Sorry. This is an invalid entry. Please try again.") #notifies user if input was invalid
        main_function() #allows user to try again
    print("Today's weather forecast for ",prompt,":")
    json_response_city=response_city.json() #converts from json
    temp=json_response_city["main"]["temp"] #gets current temperature
    print("Current temperature:",temp,"F")
    high=json_response_city["main"]["temp_max"] #gets temp high
    print("High Temperature:",high,"F")
    low=json_response_city["main"]["temp_min"] #gets temp low
    print("Low Temperature:",low,"F")
    humidity=json_response_city['main']['humidity'] #gets humidity
    print("Humidity: ",humidity,"%")
    weather_condition=json_response_city["weather"][0]["description"] #gets weather condition
    print("Weather Condition: ",weather_condition)
    repeat_function()

def repeat_function():
    while True:
        prompt=input("Would you like to see the weather for another area? Please type yes or no.\n") #prompt for another look up
        if prompt=="no":
            print("Have a great day!") #if no, exits code
            exit()
        elif prompt=="No":
            print("Have a great day!") #if no, exits code
            exit()
        elif prompt=="yes":
            main_function() #if yes, repeats main
        elif prompt=="Yes":
            main_function() #if yes, repeats main
        else:
            print("Invalid entry. Please try again.") #if user inputs something other than accepted responses

main_function()
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y_pos = np.arange(len(x))
y = [10,8,6,4,2,1]
```

```python
#bar graph
plt.bar(x,y)
plt.title("Bar Graph")
plt.xlabel('Languages')
plt.ylabel("Usage")
plt.show()
```

```python
#scatterplot
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel('Languages')
plt.ylabel('Usage')
plt.show()
```

```python
#piechart
plt.pie(y,labels=x,autopct='%.2f')
plt.xlabel('Languages')
plt.title("Pie Chart")
plt.show()
```
